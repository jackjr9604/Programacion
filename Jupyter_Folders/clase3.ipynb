Con el metodo value_counts, se cuenta los parametros de la serie seleccionada, aqui por ejemplo le pedimos sobre la serie Vendedor y nos da la cantidad de vendedores y la cantidad de registros que existe para cada vendedor

nos da los valores desendentes por q la propiedad ascendente esta desactivada

```python
tabla.Vendedor.value_counts()
```

si le colocamos al metodo el parametro ascending y le damos un valor true nos va a mandar la informacion en orden ascendente, y con el parametro dropna podemos desactivar y ver si existe algun valor nulo por lo q no mas nos mostrara los registros q no sean nulos

```python
tabla.Región.value_counts(ascending=True, dropna=False)
```

el parametro sourt_values le da a un objeto un ordenamiento, en este caso le dio a la seri vendedor un orden alfabeltico y como resultado nos imprime los primeros y los ultimos mostrando el orden que tiene

```python
tabla.Vendedor.sort_values()
```

a un objeto con el parametro by= y seleccionando las series a ordenar imprime los registros ordenados desendentemente, en este caso imprime la serie vendedor en orden alfabetico y cada registro de vendedor lo organiza segun el valor de la ganancia descendentemente

```python
tabla.sort_values(by=['Vendedor', 'Ganancia'])
```

con la funcion == podemos imprimir la informacion que queremos ver especificamente, de la siguente manera nos imprime un false o un truo segun el parametro que le colocamos a la serie, pero no es vuasualmente comodo

```python
tabla.Artículo == 'Gorras'
```

asi que encerrando lo anterior en dos corchetes [] y colocando el dataframe antes de los corchetes nos va a imprimir todo lo q necesitamos pero en forma de tabla para q podamos leer mejor la informacion asi como se muestra acontinuacion

```python
tabla[tabla.Artículo == 'Gorras']
```

y de la siguente forma colocando un & podemos pedir dos o mas caracteristicas para imprimir, en este caso podemos ver las gorras q se vendieron en la region del Norte

```python
tabla[(tabla.Artículo == 'Gorras') & (tabla.Región == 'Norte')]
```

podemos dar manejos de strings con diferentes metodos, en este caso podemos usar el parametro str.contains para decirle q nos ubique todas las filas en donde el vendedor tenga la palabra Trevor

```python
tabla[tabla.Vendedor.str.contains('Trevor')]
```

con el parametro strarswitch podemos buscar en una serie con la letra en la q empiza

con el metodo isnumeric podemos en un contenido mixto buscar cuales tienen numeros

```python
tabla[tabla.Vendedor.str.startswith('T')]
```

importamos la libreria matplotlib como plt, y agregamos el codigo magico para poder ver los graficos despues del codigo escrito

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

creamos una variante a la cual le asignamos la fincion de que muestre la tabla filtrando en la serie categoria todo lo que contenga la palabra comestible

```python
ali = tabla[tabla.Categoría == 'Comestibles']
```

llamamos la tabla con el parametro .head para mostrar las 5 primeras lineas

```python
ali.head()
```

y mostramos el grafico de los valores contados simplemente colocando el metodo .plot y nos va a mostrar un grafico lineal, le colocamos un ; al final para q nos borre una linea de codigo arriba, ahora con el parametro figsize podemos cambiar la medida del grafico dandoles las medidas exactas

```python
ali.Artículo.value_counts().plot(color='orange', figsize=(10,5));
```

agregando la expresion Kind= 'bar' podemos mostrar nuestra informacion en un grafico de barras

```python
ali.Artículo.value_counts().plot(kind= 'bar');
```

agregando el h al final de bar podemos ver el grafico de barras horizontal y con el parametro color puede cambiarse el color del grafico

```python
ali.Artículo.value_counts().plot(kind= 'barh', color='red');
```

con la expresion pie podemos ver nustra informacion e un grafico de torta, podemos cambiar los mapas de colores con el parametro colormap= y colocando el nombre del mapa de colores

```python
ali.Artículo.value_counts().plot(kind= 'pie', colormap= 'twilight');
```
